**Challenge 1**

In this challenge, we will be working with pokemon data. The data can be found here:

- https://raw.githubusercontent.com/data-bootcamp-v4/data/main/pokemon.csv

```python
#libraries
import pandas as pd
import scipy.stats as st
import numpy as np

```

```python
df_pokemon = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/pokemon.csv")
df_pokemon
```

- We posit that Pokemons of type Dragon have, on average, more HP stats than Grass. Choose the propper test and, with 5% significance, comment your findings.

```python
dragon_hp = df_pokemon[(df_pokemon['Type 1'] == 'Dragon') | (df_pokemon['Type 2'] == 'Dragon')]['HP']
grass_hp = df_pokemon[(df_pokemon['Type 1'] == 'Grass') | (df_pokemon['Type 2'] == 'Grass')]['HP']
```

```python
# Perform an independent two-sample t-test
t_stat, p_value = st.ttest_ind(dragon_hp, grass_hp, equal_var=False)

# Print the results
print(f"T-statistic: {t_stat}")
print(f"P-value: {p_value}")
```

```python
# I reject the null hypothesis. There is sufficient evidence to conclude that Dragon type Pokémon have, on average, higher HP stats than Grass type Pokémon
```

- We posit that Legendary Pokemons have different stats (HP, Attack, Defense, Sp.Atk, Sp.Def, Speed) when comparing with Non-Legendary. Choose the propper test and, with 5% significance, comment your findings.


```python
from statsmodels.multivariate.manova import MANOVA

# Select the relevant columns for comparison
stats_columns = ['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']

# Perform MANOVA
manova = MANOVA.from_formula('HP + Attack + Defense + Q("Sp. Atk") + Q("Sp. Def") + Speed ~ Legendary', data=df_pokemon)
manova_results = manova.mv_test()

# Print the results
print(manova_results)
```

```python
# This indicates that Legendary Pokemons have significantly different stats compared to Non-Legendary Pokemons.
# I reject the null hypothesis. This indicates that Legendary Pokémon have significantly different stats compared to Non-Legendary Pokémon.
```

**Challenge 2**

In this challenge, we will be working with california-housing data. The data can be found here:
- https://raw.githubusercontent.com/data-bootcamp-v4/data/main/california_housing.csv

```python
df2 = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/california_housing.csv")
```

**We posit that houses close to either a school or a hospital are more expensive.**

- School coordinates (-118, 34)
- Hospital coordinates (-122, 37)

We consider a house (neighborhood) to be close to a school or hospital if the distance is lower than 0.50.

Hint:
- Write a function to calculate euclidean distance from each house (neighborhood) to the school and to the hospital.
- Divide your dataset into houses close and far from either a hospital or school.
- Choose the propper test and, with 5% significance, comment your findings.
 

```python
# Function to calculate Euclidean distance
def euclidean_distance(lat1, lon1, lat2, lon2):
    return np.sqrt((lat1 - lat2)**2 + (lon1 - lon2)**2)

# School and hospital coordinates
school_coords = (-118, 34)
hospital_coords = (-122, 37)

# Calculate distances
df2['distance_to_school'] = df2.apply(lambda row: euclidean_distance(row['longitude'], row['latitude'], school_coords[0], school_coords[1]), axis=1)
df2['distance_to_hospital'] = df2.apply(lambda row: euclidean_distance(row['longitude'], row['latitude'], hospital_coords[0], hospital_coords[1]), axis=1)

df2
```

```python
# Define the threshold for being close
threshold = 0.50

# Create a new column to indicate if a house is close to either a school or a hospital
df2['close_to_school_or_hospital'] = (df2['distance_to_school'] < threshold) | (df2['distance_to_hospital'] < threshold)

# Divide the dataset into two groups
close_houses = df2[df2['close_to_school_or_hospital']]
far_houses = df2[~df2['close_to_school_or_hospital']]

# Display the results
print(f"houses close to a school or hospital: {len(close_houses)}")
print(f"houses far from a school or hospital: {len(far_houses)}")
```

```python
# Perform an independent two-sample t-test
t_stat, p_value = st.ttest_ind(close_houses['median_house_value'], far_houses['median_house_value'], equal_var=False)

# Print the results
print(f"T-statistic: {t_stat}")
print(f"P-value: {p_value}")
```

```python
#I reject the null hypothesis. Houses close to a school or hospital are significantly more expensive.
```
